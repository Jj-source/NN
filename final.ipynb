```python
# -*- coding: utf-8 -*-
"""FinalAssignment.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/11aXHwGLRsUUT72VWCfrhlATYAWH2yv35

---

**Name**: 
Gili Jacopo
---

**Student number**: 
s5431492
---
"""
##From - https://stackoverflow.com/questions/32419510/how-to-get-reproducible-results-in-keras/52897216#52897216

## INCLUDE THIS AT THE BEGINNING OF YOUR PYTHON PROGRAM TO MAKE YOUR EXPERIMENTS REPLICABLE

seed = 10
import os
os.environ['PYTHONHASHSEED']=str(seed)

# 2. Set python built-in pseudo-random generator at a fixed value
import random
random.seed(seed)

# 3. Set numpy pseudo-random generator at a fixed value
import numpy as np
np.random.seed(seed)

# 4. Set tensorflow pseudo-random generator at a fixed value
import tensorflow as tf
tf.random.set_seed(seed)

# Lets start with some functions
# I wrote this program in google colab, hope it works as a python script.

import numpy as np
import pandas as pd
import sys

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten
from tensorflow.keras.layers import SimpleRNN,LSTM,GRU
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical


# Convert List of lists to list of Strings, necessary to dataprep
def stringify(column):
  tmp = df_data[column].astype(str).values.tolist()
  test_list = tmp
  tmp = [''.join(ele) for ele in test_list]
  return tmp

# transform the string in a padded array
def dataprepSingle(stri):
  li = list(stri.split("-"))
  enc = my_tokenizer.texts_to_sequences(li)
  pad = pad_sequences(enc, maxlen=17, padding='post')
  return pad

#helps the user understand the prediction printing the country
def getCountry(prediction):
  y_pred  = np.argmax(prediction, axis=-1)
  countries = ['American', 'Belgian', 'Dutch', 'Indian', 'Italian', 'Zimbabwean']
  return countries[y_pred[0]]

#percentage of certainty of the guess. interesting
def getPercentage(prediction):
  y_pred  = np.argmax(prediction, axis=-1)
  return prediction[0][y_pred[0]] * 100

# function that makes a prediction on a singular string (a surname)
def predict(stri, mode, verbose):
  if(mode):
    model.load_weights(checkpoint_path)
  pr = model.predict(dataprepSingle(stri))
  if(verbose):
    print("prediction for " + stri + " is " + getCountry(pr) + " with ")
    print(getPercentage(pr))
  else:
    print(stri + ", " + getCountry(pr))


from sklearn.dummy import DummyClassifier
def dummy():
  x_train_dummy,x_test_dummy,y_train_dummy,y_test_dummy=train_test_split(
      df_data[feature], df_data[target],
      random_state=42,
      shuffle=True,
      test_size=0)
  my_dummy_model = DummyClassifier(strategy="most_frequent")
  my_dummy_model.fit(x_train_dummy, y_train_dummy)

  # predict and add predictions to our data frame 
  Array = my_dummy_model.predict(surnames_blind)
  #print("baseline with most frequent guess has an accuracy on test set of:")
  #print(Dummy_score)

# 1) Load dataset from .csv using pandas dataframe, if running the notebook locally

#df_data = pd.read_csv("names_culture_origin.csv", sep="\t")
df_data = pd.read_csv('https://raw.githubusercontent.com/Jj-source/NN/main/names_culture_origin.csv', sep='\t')

target = "Culture"
feature = "Surname"

#print("Dataset loaded!")


# 2)start of data preparation, 

surnames = stringify(feature)
labels = stringify(target)

# Load the blind dataset
#blind_data = pd.read_csv(sys.argv[1], sep="\t")
blind_data = pd.read_csv('https://raw.githubusercontent.com/Jj-source/NN/main/blind_names.csv', sep="\t")
surnames_blind = stringify(feature)

"""# **Embedding**"""

"""Whatever your task is (language modeling, document classification, etc.), the NN input needs to have fixed dimensions
and so we need to transform all the input sentences so they are of the same size."""

# lets look for the longest surname to calculate adeguate padding
longest_surname_dataset1 = max(surnames, key=len)
longest_surname_dataset2 = max(surnames_blind, key=len)
if(longest_surname_dataset1 > longest_surname_dataset2):
  longest_surname = longest_surname_dataset1
else:
  longest_surname = longest_surname_dataset2


my_tokenizer = Tokenizer(lower=False , char_level = True)
alphabet = " ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
my_tokenizer.fit_on_texts(alphabet)
encoded_input = my_tokenizer.texts_to_sequences(surnames)
padded_input = pad_sequences(encoded_input, maxlen = len(longest_surname), padding='post')
vocab_size = len(my_tokenizer.word_index) + 1

"""# **labels**

Now we need to format the target column ('label') as ***categorical data***, that is data which takes on a finite number of possible values.
"""

labels = df_data[target].astype('category')
nb_classes = labels.nunique()

"""You can see above that Pandas has assigned a fixed ID to each possible category, and transformed the data accordingly.

Next, the integer ID representation needs to be represented to a **one-hot representation** so it is compatible with the final layer of a (multi-class) NN.
This can be done with Keras' utility function [to_categorical](https://keras.io/api/utils/python_utils/#to_categorical-function).

The resulting matrix has a row for each sample, and the number of columns is the number of classes to be predicted. Only the item corresponding to the class ID will be set to 1, the rest to zeros.
"""

labels_onehot = to_categorical(labels.cat.codes)
labels_onehot

# [technical] reshape the 2D array as a 3D array (with last dimension=1)
seq_length =  len(longest_surname)
nb_samples = len(surnames)
padded_input = padded_input.reshape((nb_samples, seq_length, 1))

###INPUT HAS BEEN PREPARED, IT CAN BE NOW FED INTO THE MODEL
#print("input ready!")

x_train = padded_input
y_train = labels_onehot

    #test size not needed when we want to predict outside of the dataset

"""#### Step 4: Initializing model with Embedding layer

We show here a part of the code that is used to define a NN model and focus on the [Embedding](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding) layer, which will learn to convert the one-hot encodings to dense representations (word embeddings).

The arguments to the Embedding layer are:
- input_dim: Size of vocabulary
- output_dim: Size of word embedding
- input_length: Length of input sequences. This is 6 in our example, while in the case of FFNN-LMs explained in class it will be the length of the word context (n-1)
"""

# imports necessaire to callback functions
import os
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import Bidirectional
from sklearn import model_selection

checkpoint_path = "./final_model.ckpt"
cb_earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', mode="min", patience=15, restore_best_weights=True)
checkpoint = ModelCheckpoint(filepath= checkpoint_path,
                                       monitor="val_accuracy", verbose=1, save_best_only = True, mode="max")
                                       #or was it better with val_loss ?




# the model itself!!!!

size_of_embedding = 256
seq_length =  len(longest_surname)
nb_timesteps = 1     #https://amitness.com/2020/04/recurrent-layers-keras/
                                  #we are passing one character at a time, every padded input has 17 chars
nb_features = size_of_embedding   #every timesteps is composed of one char so, 1 feature per timestep
                                  #all of this would be tru if we didn do the embedding.....
                                  #in theory: we are passing one surname at a time, embedded as a 32 bit array.
                                  #I would say it is features = size of embedding and timestep = 1 
hidden_dim = 128
# add layers to network
model=Sequential()  
model.add(Embedding(input_dim=vocab_size, output_dim = size_of_embedding, input_length = len(longest_surname)))
model.add(LSTM(hidden_dim, input_shape = (nb_timesteps, nb_features), activation='tanh'))
model.add(Dense(500, activation='tanh'))
model.add(Dense(nb_classes,activation='softmax'))     #Output layer (multi-class)

model.summary()

# compile
model.compile(
  loss='categorical_crossentropy',
  optimizer='Adam',
  metrics=['accuracy'],
)

# fit with VALIDATION and EARLY STOPPING strategy
train_history = model.fit(
    x_train, y_train, epochs=1000, initial_epoch=0, verbose=1,
    validation_split=0.2,
    callbacks = [cb_earlystop, checkpoint])



df_train_history=pd.DataFrame(train_history.history)
df_train_history[['accuracy','val_accuracy']].rolling(10).mean().plot()

#here we will make the new predictions

for surname in surnames_blind:
  predict(surname, 0, 0)

print("\n\n-----")
print("with loaded weights from checpoint save with 'max validation accuracy")

for surname in surnames_blind:
  predict(surname, 1, 0)

#butttt we want to speriment with the checkpoint value, so I'll make also prediction with that
#saved weights.. for the competition the first set of predictions can be used. I am just
#curious about the outcome
```
